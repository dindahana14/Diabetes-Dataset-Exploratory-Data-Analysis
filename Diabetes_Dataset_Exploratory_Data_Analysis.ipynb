### Identifikasi outlier dari kolom `Glucose` berdasarkan mean dan standar deviasi menggunakan `3-Sigma Rule`



```python
mean = df[['Glucose']].mean()
stdev = df[['Glucose']].std()
iso = (df[['Glucose']] < mean - 3*stdev) | (df[['Glucose']] > mean + 3*stdev)
df['is_outlier'] = iso

df1 = df[['Glucose', 'Outcome']].assign(is_outlier=iso)
df1.loc[df1['is_outlier']]
```

- Outlier pada Outcome 1: Ditemukan 2 outlier dalam kelompok yang memiliki diabetes. Hal ini menunjukkan adanya nilai-nilai yang sangat jauh dari sebagian besar data dalam kelompok ini. Outlier dalam kelompok yang memiliki diabetes dapat menunjukkan variasi yang ekstrem dalam beberapa fitur yang mungkin menjadi indikasi adanya kasus diabetes yang lebih parah atau tidak biasa.
- Outlier pada Outcome 0: Ditemukan 3 outlier dalam kelompok yang tidak memiliki diabetes. Hal ini juga menunjukkan adanya nilai-nilai yang sangat jauh dari sebagian besar data dalam kelompok ini. Outlier dalam kelompok yang tidak memiliki diabetes mungkin menunjukkan variasi ekstrem dalam karakteristik individu yang mungkin mempengaruhi risiko terhadap diabetes atau faktor-faktor lainnya yang relevan.

### Identifikasi extreme outlier dari kolom `Glucose` berdasarkan quartil pertama, ketiga, dan range antar kedua kuartil tersebut menggunakan *`Tukey's Fences`*



```python
q1 = df['Glucose'].quantile(q=0.25)
q3 = df['Glucose'].quantile(q=0.75)
iqr = q3 - q1
iso = (df['Glucose'] < q1 - 1.5*iqr) | (df['Glucose'] > q3 + 1.5*iqr)
iseo = (df['Glucose'] < q1 - 3*iqr) | (df['Glucose'] > q3 + 3*iqr)
df1 = df[['Glucose', 'Outcome']].assign(is_outlier = iso, is_extreme_outlier = iseo)
df1.loc[df1['is_outlier'] | df1['is_extreme_outlier']]
```

- Tidak ditemukan outlier yang extreme pada kedua outcome yang menunjukkan tidak ada nilai Glukosa yang secara signifikan jauh dari nilai-nilai lainnya dalam kelompok tersebut. Hal ini menunjukkan bahwa distribusi Glukosa mungkin relatif homogen atau tidak memiliki nilai ekstrem yang sangat jauh dari nilai-nilai lainnya.
